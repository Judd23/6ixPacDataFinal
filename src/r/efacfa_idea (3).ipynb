```r
# ------------------------------
# 1. Install & load required packages
# ------------------------------
install.packages(c("psych", "lavaan", "semTools", "GPArotation", "dplyr", "ggplot2"))

library(psych)
library(lavaan)
library(semTools)
library(GPArotation)
library(dplyr)
library(ggplot2)

# ------------------------------
# 2. Load dataset
# ------------------------------
data <- read.csv("/content/OfficialDataset_Final (1).csv")  # replace with your file path

# ------------------------------
# 3. Define latent factor variables
# ------------------------------
F1_vars <- c("FATHEDUC","MOTHEDUC","RACEGROUP","UndrRepStud_RC","FINCON_RC","FirstGen_RC",
             "INCOME","HSGPA","CITIZEN","PLANLIVE","DSstud_RC","SIFRAT01","SIFRAT07","SIFMAJA")

F2_vars <- c("COLACT17","COLACT19","SATIS01","SATIS07","SATIS02","SATIS13","SATIS15","SATIS25",
             "CSSRAT16","CSSRAT07","GENACT05","CSSHPW01","CSSHPW02","CSSHPW03","CSSHPW05","CSSHPW08")

F3_vars <- c("SLFCHG01","SLFCHG02","SLFCHG03","SLFCHG04","CSSRAT01","CSSRAT07","SUCCESS4","SATIS13")

F4_vars <- c("SLFCHG01","SLFCHG02","SLFCHG03","SLFCHG04","SATIS13")

all_items <- data %>% select(all_of(unique(c(F1_vars,F2_vars,F3_vars,F4_vars))))

# ------------------------------
# 4. Exploratory Factor Analysis (EFA)
# ------------------------------
# Parallel analysis
fa.parallel(all_items, fa="fa", n.iter=100, main="Parallel Analysis")

# EFA: 4 factors, oblimin rotation
efa <- fa(all_items, nfactors=4, rotate="oblimin", fm="pa")
print(efa$loadings, cutoff=0.40)

# Automatically select items with loadings >= 0.40 on any factor
high_load_items <- rownames(efa$loadings)[apply(abs(efa$loadings[]), 1, max) >= 0.40]
all_items_clean <- all_items[, high_load_items]

# ------------------------------
# 5. Dynamically create CFA model
# ------------------------------
F1_model_vars <- intersect(F1_vars, high_load_items)
F2_model_vars <- intersect(F2_vars, high_load_items)
F3_model_vars <- intersect(F3_vars, high_load_items)
F4_model_vars <- intersect(F4_vars, high_load_items)

cfa_model <- paste0(
  "F1 =~ ", paste(F1_model_vars, collapse=" + "), "\n",
  "F2 =~ ", paste(F2_model_vars, collapse=" + "), "\n",
  "F3 =~ ", paste(F3_model_vars, collapse=" + "), "\n",
  "F4 =~ ", paste(F4_model_vars, collapse=" + ")
)

cat("CFA model syntax:\n")
cat(cfa_model)

# ------------------------------
# 6. Run CFA
# ------------------------------
cfa_fit <- cfa(cfa_model, data=all_items_clean, estimator="MLR")
summary(cfa_fit, fit.measures=TRUE, standardized=TRUE, rsquare=TRUE)
fitMeasures(cfa_fit, c("cfi","tli","rmsea","srmr"))
reliability(cfa_fit)

# Optional: measurement invariance
# measurementInvariance(model=cfa_model, data=all_items_clean, group="GENDER")

# ------------------------------
# 7. Dynamically create SEM model
# ------------------------------
sem_model <- paste0(
  "F1 =~ ", paste(F1_model_vars, collapse=" + "), "\n",
  "F2 =~ ", paste(F2_model_vars, collapse=" + "), "\n",
  "F3 =~ ", paste(F3_model_vars, collapse=" + "), "\n",
  "F4 =~ ", paste(F4_model_vars, collapse=" + "), "\n",
  "F2 ~ F1\n",
  "F3 ~ F1 + F2\n",
  "F4 ~ F1 + F2 + F3"
)

cat("\nSEM model syntax:\n")
cat(sem_model)

# ------------------------------
# 8. Run SEM
# ------------------------------
sem_fit <- sem(sem_model, data=all_items_clean, estimator="MLR")
summary(sem_fit, fit.measures=TRUE, standardized=TRUE, rsquare=TRUE)
fitMeasures(sem_fit, c("cfi","tli","rmsea","srmr"))
```
